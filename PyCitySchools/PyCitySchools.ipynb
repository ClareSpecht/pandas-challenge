```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate total number of schools
schools = len(school_data)

#Calculate total number of students
students = len(student_data)

#Calculate total budget
budget = school_data["budget"].sum()

#Calculate average math score
mathscore = student_data["math_score"].mean()

#Calculate average reading score
readscore = student_data["reading_score"].mean()

#Calculate percentage of students with passing math score
mathpass = len(student_data.loc[student_data["math_score"] >= 70])
percentmathpass = mathpass / students * 100

#Calculate percentage of students with passing reading score
readpass = len(student_data.loc[student_data["reading_score"] >= 70])
percentreadpass = readpass / students * 100

#Calculate percentage of students with passing math and reading scores
overallpass = len(student_data.loc[(student_data["reading_score"] >= 70) & (student_data["math_score"] >= 70)])
percentoverallpass = overallpass / students * 100

#Create dataframe to hold results
districtsum = pd.DataFrame({"Total Schools" : [schools], "Total Students" : [students], "Total Budget" : budget, 
                            "Average Math Score" : mathscore, "Average Reading Score" : readscore, "% Passing Math" : 
                            percentmathpass, "% Passing Reading" : percentreadpass, "% Overall Passing" : 
                            percentoverallpass})

#Format data
districtsum["Total Students"] = districtsum["Total Students"].map("{:,.0f}".format)
districtsum["Total Budget"] = districtsum["Total Budget"].map("${:,.2f}".format)
districtsum
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Rename column in case we'd like index name to show up
newschool = school_data.rename(columns={"school_name" : "School Name"})
newstudent = student_data.rename(columns={"school_name" : "School Name"})

#Group by school name
groupedschool = newschool.groupby(["School Name"])
groupedstudent = newstudent.groupby(["School Name"])

#Generate school type per school
schooltype = groupedschool["type"].unique()

#Pull school type out of list generated by unique command
for x in range(len(schooltype)):
    array = schooltype[x]
    schooltype[x] = array[0]

#Count students at each school
studentcount = groupedstudent["student_name"].count()

#Calculate budget at each school
budgetsum = groupedschool["budget"].sum()

#Calculate budget per student at each school
perstudent = budgetsum / studentcount

#Calculate average math score per school
schoolmath = groupedstudent["math_score"].mean()

#Calculate average reading score per school
schoolread = groupedstudent["reading_score"].mean()

#Calculate percentage of students with passing math score
schoolmathpass = newstudent.loc[newstudent["math_score"] >= 70]
groupedmath = schoolmathpass.groupby(["School Name"])
mathpassschool = ((groupedmath["math_score"]).count() / studentcount) * 100

#Calculate percentage of students with passing reading score
schoolreadpass = newstudent.loc[newstudent["reading_score"] >= 70]
groupedread = schoolreadpass.groupby(["School Name"])
readpassschool = ((groupedread["reading_score"]).count() / studentcount) * 100

#Calculate percentage of students with passing overall score
schooloverpass = newstudent.loc[(newstudent["reading_score"] >= 70) & (student_data["math_score"] >= 70)]
groupedover = schooloverpass.groupby(["School Name"])
overpassschool = ((groupedover["reading_score"]).count() / studentcount) * 100

#Create dataframe to hold results
schoolsummary = pd.DataFrame({"School Type" : schooltype,
                              "Total Students" : studentcount,
                              "Total School Budget" : budgetsum, 
                              "Per Student Budget" : perstudent,
                              "Average Math Score" : schoolmath, 
                              "Average Reading Score" : schoolread, 
                              "% Passing Math" : mathpassschool, 
                              "% Passing Reading" : readpassschool, 
                              "% Overall Passing" : overpassschool})

#Use to hide "School Name" to match starter file
schoolsummary.index.name = None

#Makes a nonformatted copy of dataframe
schoolsum = schoolsummary.copy()

#Format dataframe
schoolsummary["Total School Budget"] = schoolsummary["Total School Budget"].map("${:,.2f}".format)
schoolsummary["Per Student Budget"] = schoolsummary["Per Student Budget"].map("${:,.2f}".format)
schoolsummary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Define lower and upper limits for bins
spendbins = [0, 585, 630, 645, 680]

#Define bin names
binranges = ["<$585", "$585-630", "$630-645", "$645-680"]

#Create a column that assigns appropirate bin to every row
schoolsum["Spending Ranges (Per Student)"] = pd.cut(schoolsum["Per Student Budget"], spendbins, labels=binranges, include_lowest=True)

#Keep necessary columns
summaryshort = schoolsum[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing", "Spending Ranges (Per Student)"]]

#Group by Spending Ranges
schoolspend = summaryshort.groupby("Spending Ranges (Per Student)")

#Output averages of each metric
avgschoolspend = schoolspend.mean()
avgschoolspend.round(2)
```

## Scores by School Size
